```python
class NodoBinario:
  def __init__( self, value, left=None, right=None):
    self.data = value
    self.left = left
    self.right = right

class NodoTernario:
  def __init__( self, value, left=None, center=None, right=None):
    self.data = value
    self.left = left
    self.center = center
    self.right = right
    
```
